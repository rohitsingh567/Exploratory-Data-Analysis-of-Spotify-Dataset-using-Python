# Data Visualization

```python
# Checking relation between variables
corr_df = df2.drop(["key","mode","explicit"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,5))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="magma", linewidths="0.5")
heatmap.set_title("Correlation Heatmap Between Variables")
plt.show()
```

```python
# taking sample from large dataset
sample_df = df2.sample(int(0.004*len(df2)))
len(sample_df)
```

```python
# checking the relationship between energy and loudness
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="loudness", x="energy", color="c", marker="*").set(title="Loudness vs Energy")
plt.show()
```

```python
# checking the relationship between energy and loudness
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="popularity", x="acousticness", color="g", marker="*").set(title="Popularity vs Acousticness")
plt.show()
```

```python
# getting the years for visulization
df2["dates"] = df2.index.get_level_values("release_date")
df2.dates = pd.to_datetime(df2.dates)
years = df2.dates.dt.year
```
